```python
from pywebcopy import save_webpage

url = "https://shinrakuen-teatime.com/index.html"

kwargs = {'project_name': 'some-fancy-name'}

save_webpage(
    url=url,
    project_folder='./downloads_sample1',
    **kwargs
)
```

```python
from pywebcopy import save_website

url = "https://shinrakuen-teatime.com/index.html"

save_website(
url=url,
project_folder='./downloads_sample2',
project_name="my_site",
bypass_robots=True,
debug=True,
open_in_browser=True,
delay=None,
threaded=False,
)
```
